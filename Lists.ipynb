# Lists
1.   Ordered Sequence of data
2.   It can hold elements of different data types
3.   Collection of Values
4.   Lists are mutable in nature


```python
sample_list = ["hello world", 10, 50.36, 'A']
print(sample_list)
sample_list[3] = "text"
print(sample_list)
```

# List Functions

## Append function - to add a new element to the existing list

```python
# Lists are mutable with append function
sample_list.append(["Bread", "butter"])
print(sample_list)
```

## Count

```python
# Returns the no of times, the search element appear in the list
print(sample_list.count("hello world"))
```

## Length of the list

```python
print("Total no of elements in the list : ", len(sample_list))
```

## Sort

```python
# Sort is inplace - mutable
numeric_list = [9,7,1,5,6,0]
numeric_list.sort()
print(numeric_list)
```

```python
numeric_list.sort(reverse = True)
print(numeric_list)
```

```python
string_list = ["hello", "how", "are", "you", "world"]
string_list.sort()
print(string_list)
```

## Remove an element from the list

```python
string_list.remove("you")
print(string_list)
```

## Clear

```python
string_list.clear()
print(string_list)
```

## List - Point by reference

```python
list1 = [2,3,4]
list2 = list1
list2.pop()
print(list1)
print(list2)
```

## List Copy

```python
list1 = [1,2,3]
list2 = list1.copy()
# Insert takes two parameters - index, value to be inserted
list1.insert(3, 4)
print(list1)
print(list2)
```

## Reverse

```python
list1 = [8,9,7,9,0]
list1.reverse()
print(list1)
```

## Extend

```python
list1 = [1,2,3,4,5]
list1.extend(list1)
print(list1)
```
