# Tuples


1.   Ordered Sequence of items
2.   Holds items of different data types
3.   Immutable in nature



```python
sample_tuple = ("a", "text", "a", 1, 1000090, 90.89)
print(sample_tuple)
```

# Tuple Functions

## Count

```python
print(sample_tuple.count("a"))
```

## Index

```python
print(sample_tuple.index(1000090))
```
