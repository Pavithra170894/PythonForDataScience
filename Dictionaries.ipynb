## Dictionaries


1.   Unordered Collection of key value pairs
2.   Keys are unique and immutable
3.   Keys are case sensitive




```python
my_dict = {
    "name" : "Pavithra",
    "age"  : 28,
    3      : 10000
}
```

```python
print(my_dict)
```

# Items - returns each key value pair as tuple

```python
print(my_dict.items())
```

```python
for key,value in my_dict.items():
    print("key : ", key, " value: ", value)
```

# Values are mutable in nature in dictionary

```python
my_dict[3] = "test"
```

```python
print(my_dict)
```

# Get only keys of the dictionary

```python
print(my_dict.keys())
print("VALUES : ")
for key in my_dict.keys():
    print(my_dict[key])
```

## Get only values from the dictionary

```python
print(my_dict.values())
```

## Copy

```python
new_dict = my_dict.copy()
new_dict["age"] = 27
print("New_dict : ", new_dict)
print("Old dict : ", my_dict)
```

# Pop and PopItem

```python
print(new_dict.popitem())
```

```python
print(new_dict.pop("age"))
```

# Get an item from the dictionary

```python
print(my_dict.get("name", "age"))
```

```python
# Second paramter is the default value if key not found
print(my_dict.get("test", "age"))
```

# Update the dictionary

```python
my_dict.update({"city":"chennai"})
print(my_dict)
```

# Clear the dictionary

```python
my_dict.clear()
print(my_dict)
```
