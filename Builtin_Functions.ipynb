# Max, Min and Len functions

```python
x = [10, 7, 6, 15, 0, 14]

print("Array : ", x)
print("Maximum value of the array : ", max(x))
print("Minimum value of the array : ", min(x))
print("Total elements in the array : ", len(x))
```

# Sort Function

```python
x = [0, 6, 3, 8, 10, 4, 2, 8]
print(sorted(x))
print(sorted(x, reverse=True))
print(x)
```

# Abs function

```python
x = -1
x1 = 1
print(abs(x))
if abs(x) == x1:
    print("Both variables are same")
```

# Floor, Ceil function

```python
import math

x = 5.6
print(math.floor(x))
print(math.ceil(x))
```
