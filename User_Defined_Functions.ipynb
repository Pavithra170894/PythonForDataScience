# Functions in python

```python
def test():
    print("Function without any paramerts or return function")

test()
```

```python
def test(name, ):
    print("Function with one parameter and no return function")
    print(name)

test("pavithra")
```

```python
def test(name, age=18):
    print("Function with one parameter and another parameter with default value and no return function")
    print(name, age)

test("pavithra")
```

```python
def test(name, age=18, *args):
    print("Function with variable number of paramertes and no return function")
    print(name, age)
    print(args[0])
    print(args[1])

test("pavithra", 28, [10,20,40], (10,20))
```

```python
def test(*args):
    print("Function with variable number of paramertes and a return function")
    return args[0] * args[1]

result = test(10, 20)
print(result)
```
